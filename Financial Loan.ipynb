# Exploratory Data Analysis(EDA)

```python
plt.figure(figsize=(10,6))
plt.bar(loan['total_acc'],loan['total_payment'],color='blue')
plt.title('Total Payment of Home Owner ship')
plt.xlabel('home_ownership')
plt.ylabel('total_payment')
plt.show
```

```python
correlation=loan.corr()
correlation
```

```python
sns.heatmap(correlation,annot=False,cmap='coolwarm')
```

```python
loan['verification_status'].value_counts().plot(kind='pie',autopct='%1.0f%%')
```
